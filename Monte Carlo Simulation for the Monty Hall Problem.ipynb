# Using a Monte Carlo Simulation to solve the Monty Hall Problem

The Monty Hall problem is a famous problem in probability, named after the American game show host. It takes on the following format: A contestant on a gameshow has to choose between one of three doors. One of the doors has a car behind it, whereas the other two have goats. The contestant's goal is to choose the door with the car behind it. The game starts with all the doors closed and the contestant choosing one of the doors. Monty then opens a door with a goat behind it and offers the contestant the opportunity to switch their chosen door to the other reamining closed door or to stick with their initial choice. What should the contestant do? Intuitively they have a 50% chance of choosing the correct door whether they decide to stick or switch. However, appearances can be deceptive and this is incorrect. The contestant actually has a $2/3$ chance of winning if they switch doors and only a $1/3$ chance of winning if they decide to stick.

Below we demonstrate how the Monty Hall problem can be solved with a Monte Carlo simulation, and also with Bayes' theorem.

```python
import random


def monte_carlo(no_trials):

    stick_wins = 0
    switch_wins = 0

    for trial in range(no_trials):

        # the contestant chooses a door at random from the three available options
        contestant_door = random.randint(1, 3)
        # the car is randomly placed behind one of the three doors
        car_door = random.randint(1, 3)
        # Monty has to open one of the doors with a goat behind, so he has to open a door
        # that isn't the contestant's door and doesn't have the car behind
        monty_options = [door for door in [1, 2, 3] if door != contestant_door and door != car_door]
        # Monty randomly chooses a door from the available options
        # if the contestant has chosen the door with the car behind then Monty has two options
        # to choose from. If not, he only has one available option
        monty_door = random.choice(monty_options)
        # if the contestant chooses to switch, they must switch to the door that isn't their original
        # pick and isn't the door opened by Monty
        switch_door = [door for door in [1, 2, 3] if door != contestant_door and door != monty_door][0]

        if contestant_door == car_door:
            stick_wins += 1
        if switch_door == car_door:
            switch_wins += 1

    stick_wins_pct = stick_wins / no_trials
    switch_wins_pct = switch_wins / no_trials

    print("Wins when the contestant doesn't switch doors: {:.2f}%".format(stick_wins_pct*100))
    print("Wins when the contestant switches: {:.2f}%".format(switch_wins_pct*100))
```

```python
monte_carlo(100000)
```

## Solving the Monty Hall problem with Bayes' theorem

The car has the same probability of being behind each of the three doors:
$$P(car_1)=P(car_2)=P(car_3)=\frac{1}{3}$$
The contestant chooses door number 1. Monty can open either door 2 or door 3. By Bayes' theorem, the probability that the car is behind door 1 given that Monty has opened door 3 is
$$P(car_1|open_3)=\frac{P(open_3|car_1)P(car_1)}{P(open_3)}$$
$P(open_3)$ is given by 
$$\begin{align}P(open_3) &= \sum_i P(open_3|car_i)P(car_i) \notag\\
&= P(open_3|car_1)P(car_1)+P(open_3|car_2)P(car_2)+P(open_3|car_3)P(car_3) \notag\\
&= \frac{1}{2}\cdot\frac{1}{3}+1\cdot\frac{1}{3}+0 \notag\\
&= \frac{1}{2} \notag
\end{align}
$$

Note that $P(open_3|car_2)=1$ since the contestant has chosen door 1 and the car is behind door 2, so Monty can only open door 3. Also, $P(open_3|car_3)=0$ since Monty cannot open a door with the car behind.

Therefore, we have
$$P(car_1|open_3)=\frac{\frac{1}{2}\cdot\frac{1}{3}}{\frac{1}{2}}=\frac{1}{3}$$
This means we also have
$$P(car_2|open_3)=1-P(car_1|open_3)=\frac{2}{3}$$

So, if the contestant initially chooses door 1 and Monty opens door 3, the contestant has a $1/3$ chance of winning the car if they stick with door 1, and a $2/3$ chance of winning if they switch to door 2.
